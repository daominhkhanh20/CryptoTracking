```python
from src.db.mongodb import MongoClientConnector
from src.utils.io import load_yaml_file
config = load_yaml_file('../config/config.yml')
```

```python
mongodb = MongoClientConnector(uri_mongo_connect=config['environments']['uri_mongodb_connect'])
```

```python
tokenomics = mongodb.get_all_documents(database='information', collection='tokenomic')
```

```python
explorers = []
for token in tokenomics:
    explorers.extend(token['urls']['explorer'])
```

```python
len(explorers)
```

```python
len(set(explorers))
```

```python
filters = ['bscscan', 'ethscan']
explorer_filters = []
for ex in explorers:
    if any([word in ex for word in filters]):
        explorer_filters.append(ex)
```
